```python
#결과
import queue
import copy

Owin_n = 0
Xwin_n = 0
draw_n = 0
    

def win(board, player):
    #가로
    for i in range(3):
        #3줄이니까
        count = 0
        for j in range(3):
           
            if board[i][j] == player:
                count += 1               
        if count == 3:
            return True
        
    #세로
    for j in range(3):
        #3줄이니까 
        count = 0
        for i in range(3):
            if board[i][j] == player:
                count += 1
        if count == 3:
            return True
                
    #왼 - 오 대각선
    #하나의 경우만 있으므로
    count = 0
    for i in range(3):
        if board[i][i] == player:
            count += 1
    if count == 3:
        return True

    #오 - 왼 대각선
    #하나의 경무만 있으므로
    count = 0
    for i in range(3):
        if board[i][2-i] == player:
            count += 1   
    if count == 3:
        return True

def draw(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '':
                return False #비어있는 칸이 하나라도 있으면 그건 비긴 게 아니기에 False
    return True

def BFS(board,start_player):
    global Owin_n, Xwin_n, draw_n
    
    #큐 제작
    que = queue.Queue()
    #queue에 board와 start_player 넣기
    que.put((board,start_player))
    
    while not que.empty():
        #비어있지 않을 때까지 실행
        current_board, current_player = que.get()
        #current_board, current_player에 큐에 들어간 값을 빼낸다
        
        if win(current_board, 'O'):
            Owin_n += 1
        elif win(current_board, 'X'):
            Xwin_n += 1
        elif draw(current_board):
            draw_n += 1
        else:
            for i in range(3):
                for j in range(3):
                    if current_board[i][j] == '':
                        #기호가 없는 빈칸이라면 current_board를 복사해서 재생성
                        #즉, 깊은 복사를 해야됨                        
                        next_board = []
                        next_board = copy.deepcopy(current_board)
                        #깊은 복사 후에 기호 지정
                        next_board[i][j] = current_player
                        
                        if current_player == 'O':
                            next_player = 'X'
                        else:
                            next_player = 'O'
                            
                        que.put((next_board, next_player))
                        #큐에 값을 넣는다.
my_board = [['','',''],
            ['','',''],
            ['','','']]

BFS(my_board, 'O')

sum_num = Owin_n + Xwin_n + draw_n

print("중간에 생성되는 상태의 총 수 : %d개" % sum_num)
print("O가 이기는 경우의 수 : %d개" % Owin_n)
print("X가 이기는 경우의 수 : %d개" % Xwin_n)
print("비기는 경우의 수 : %d개" % draw_n)
```
