### 오늘의 연습문제
### 7주차 3층 신경망(4.16)에서 설계된 네트워크
### 10주차 3층 신경망은 유사한 설계를 가져야합니다.
### 첫번째 입력층 노드 3개
### 두번째 은닉층 노드 2개
### 세번째 출력층 노드 2개
### W3, B3는 출력의 개수와 동일
### 이 구성은 10주차 3층 신경망 설계도와 뼈대가 같아야한다.
### 입력의 개수와 은닉층의 노드 개수가 다르지만 뼈대가 일치해야한다

```python
#7주차
import numpy as np
import random

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

def init_network():
    network = {}
    network['W1'] = np.random.rand(3,3)

    print("w1의 랜덤값은 :\n", network['W1'])

    #6개 2,3

    network['b1'] = np.random.rand(1,3)
    print("b1의 랜덤값은 :\n", network['b1'])

    #3개 1,3
    network['W2'] = np.random.rand(3,2)
    print("W2의 랜덤값은 :\n", network['W2'])

    #6개 3,2
    network['b2'] = np.random.rand(1,2)
    print("b2의 랜덤값은 :\n", network['b2'])
    #2개 1,2
    network['W3'] = np.random.rand(2,2)
    print("W3의 랜덤값은 :\n", network['W3'])
    #4개 2,2
    return network

#입력값 3개, 1번째 은닉층 노드 3개, 2번째 은닉층 노드 2개,  출력값 2개

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2 = network['b1'], network['b2']
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3)
    y = softmax(a3)
    return y

network = init_network()
x = np.random.rand(1,3)
y = forward(network, x)
print("결과는 :", y)
```
